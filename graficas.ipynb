```python
import matplotlib.pyplot as plt
import numpy as np

# Coordenadas dispersas y fijas para aparcamientos (rojos)
x_aparcamientos = [120, 230, 540, 760, 980, 310, 410, 620, 150, 860,
                   275, 470, 690, 840, 510, 660, 135, 935, 370, 725]
y_aparcamientos = [880, 670, 520, 300, 450, 120, 780, 640, 220, 190,
                   900, 740, 110, 510, 630, 360, 470, 820, 260, 390]

# Coordenadas dispersas y fijas para estaciones (azules)
y_estaciones = [95, 205, 365, 485, 580, 645, 715, 805, 875, 960]
x_estaciones = [100, 340, 760, 180, 610, 270, 820, 490, 150, 700]

# Crear la gráfica
plt.figure(figsize=(10, 10))
plt.scatter(x_aparcamientos, y_aparcamientos, color='red', label='Aparcamientos Bicicletas')
plt.scatter(x_estaciones, y_estaciones, color='blue', label='Estaciones Polución')

# Etiquetar aparcamientos
for i in range(len(x_aparcamientos)):
    plt.text(x_aparcamientos[i], y_aparcamientos[i] + 10, str(i + 1), color='red', fontsize=9, ha='center')

# Etiquetar estaciones
for i in range(len(x_estaciones)):
    plt.text(x_estaciones[i], y_estaciones[i] + 10, str(i + 1), color='blue', fontsize=9, ha='center')

# Trazar línea de cada estación al aparcamiento más cercano
for i in range(len(x_estaciones)):
    est_x = x_estaciones[i]
    est_y = y_estaciones[i]

    # Calcular distancias euclidianas a todos los aparcamientos
    distancias = [np.hypot(est_x - x, est_y - y) for x, y in zip(x_aparcamientos, y_aparcamientos)]
    idx_min = np.argmin(distancias)

    # Coordenadas del aparcamiento más cercano
    aparc_x = x_aparcamientos[idx_min]
    aparc_y = y_aparcamientos[idx_min]

    # Dibujar línea
    plt.plot([est_x, aparc_x], [est_y, aparc_y], color='gray', linestyle='--', linewidth=1)

# Configurar límites del plano
plt.xlim(0, 1000)
plt.ylim(0, 1000)
plt.title('Mapa con Conexiones: Estaciones a Aparcamientos Más Cercanos')
plt.xlabel('Latitud')
plt.ylabel('Longitud')
plt.grid(True)
plt.legend()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Coordenadas de aparcamientos (rojos)
x_aparcamientos = [120, 230, 540, 760, 980, 310, 410, 620, 150, 860,
                   275, 470, 690, 840, 510, 660, 135, 935, 370, 725]
y_aparcamientos = [880, 670, 520, 300, 450, 120, 780, 640, 220, 190,
                   900, 740, 110, 510, 630, 360, 470, 820, 260, 390]

# Coordenadas de estaciones (azules)
y_estaciones = [95, 205, 365, 485, 580, 645, 715, 805, 875, 960]
x_estaciones = [100, 340, 760, 180, 610, 270, 820, 490, 150, 700]

# Coordenada del punto verde
punto_verde = (400, 500)

# Crear la gráfica
plt.figure(figsize=(10, 10))
plt.scatter(x_aparcamientos, y_aparcamientos, color='red', label='Aparcamientos')
plt.scatter(x_estaciones, y_estaciones, color='blue', label='Estaciones')
plt.scatter(*punto_verde, color='green', label='Usuario', s=100, zorder=5)

# Mostrar coordenadas del punto verde
plt.text(punto_verde[0], punto_verde[1] + 15, f"({punto_verde[0]}, {punto_verde[1]})", color='green', fontsize=10, ha='center')


# Etiquetar aparcamientos
for i in range(len(x_aparcamientos)):
    plt.text(x_aparcamientos[i], y_aparcamientos[i] + 10, str(i + 1), color='red', fontsize=9, ha='center')

# Etiquetar estaciones
for i in range(len(x_estaciones)):
    plt.text(x_estaciones[i], y_estaciones[i] + 10, str(i + 1), color='blue', fontsize=9, ha='center')

# 🔁 Trazar línea de cada aparcamiento a la estación más cercana
for i in range(len(x_aparcamientos)):
    aparc_x = x_aparcamientos[i]
    aparc_y = y_aparcamientos[i]

    # Calcular distancias a todas las estaciones
    distancias = [np.hypot(aparc_x - x, aparc_y - y) for x, y in zip(x_estaciones, y_estaciones)]
    idx_min = np.argmin(distancias)

    # Coordenadas de la estación más cercana
    est_x = x_estaciones[idx_min]
    est_y = y_estaciones[idx_min]

    # Dibujar línea
    plt.plot([aparc_x, est_x], [aparc_y, est_y], color='gray', linestyle='--', linewidth=1)

# 🔁 Línea del punto verde al aparcamiento más cercano
distancias_verde = [np.hypot(punto_verde[0] - x, punto_verde[1] - y) for x, y in zip(x_aparcamientos, y_aparcamientos)]
idx_aparc_cercano = np.argmin(distancias_verde)
aparc_cercano_x = x_aparcamientos[idx_aparc_cercano]
aparc_cercano_y = y_aparcamientos[idx_aparc_cercano]

# Dibujar línea en verde
plt.plot([punto_verde[0], aparc_cercano_x], [punto_verde[1], aparc_cercano_y], color='green', linewidth=2, linestyle='-.')

# Configurar límites y leyenda
plt.xlim(0, 1000)
plt.ylim(0, 1000)
plt.title('Mapa con Conexiones: Aparcamientos y Usuario')
plt.xlabel('Latitud')
plt.ylabel('Longitud')
plt.grid(True)
plt.legend()
plt.show()
```
